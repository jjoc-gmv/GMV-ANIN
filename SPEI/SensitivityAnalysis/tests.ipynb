```python
import xarray as xr
import matplotlib.pyplot as plt
from scipy.stats import gaussian_kde
import seaborn as sns
import pandas as pd


import numpy as np
```

#### Priestley-Taylor vs Penman-Monteith

```python
# Carga el archivo NetCDF de Priestley-Taylor
archivo_nc = '//gmvstorage.gmv.es/storage/anin/SPEI/SensitivityAnalysis/SPEI3vPriestley-Taylor.nc'  # Reemplaza con la ruta de tu archivo NetCDF
dataset = xr.open_dataset(archivo_nc)

variable = dataset['__xarray_dataarray_variable__'][3]

# Extrae los datos de la variable
data = variable.values
data_flat = data.flatten()
# Limpia los datos eliminando NaN e infinitos
data_flat_cleaned_PT = data_flat[~np.isnan(data_flat) & ~np.isinf(data_flat)]

# Calcula la media de los datos limpios
media_PT = np.nanmean(data_flat_cleaned_PT)
sd_PT = np.nanstd(data_flat_cleaned_PT)
```

```python
# Carga el archivo NetCDF de Priestley-Taylor
archivo_nc = '//gmvstorage.gmv.es/storage/anin/SPEI/outputs/ANIN_SPEI3_10km_SouthAfrica_1980-2023.nc'  # Reemplaza con la ruta de tu archivo NetCDF
dataset = xr.open_dataset(archivo_nc)

variable = dataset['__xarray_dataarray_variable__'][3]

# Extrae los datos de la variable
data_base = variable.values
data_flat_base = data.flatten()
# Limpia los datos eliminando NaN e infinitos
data_flat_cleaned_base = data_flat_base[~np.isnan(data_flat_base) & ~np.isinf(data_flat_base)]

# Calcula la media de los datos
media_base = np.nanmean(data_flat_cleaned_base)
sd_base = np.nanstd(data_flat_cleaned_base)
```

```python
# Estima la densidad de probabilidad
kde = gaussian_kde(data_flat_cleaned_PT)
x_vals = np.linspace(data_flat_cleaned_PT.min(), data_flat_cleaned_PT.max(), 1000)
y_vals = kde(x_vals)

# Estima la densidad de probabilidad para la segunda variable
kde_base = gaussian_kde(data_flat_cleaned_base)
x_vals_base = np.linspace(data_flat_cleaned_base.min(), data_flat_cleaned_base.max(), 1000)
y_vals_base = kde_base(x_vals_base)

# Crea y muestra el grfico de densidad con ambas lneas y leyenda con medias y SD en lneas separadas
plt.plot(x_vals, y_vals, color='lightsalmon', label=f'Priestley-Taylor equation\nMean: {media_PT:.2f}\nSD: {sd_PT:.2f}', linewidth=3)
plt.plot(x_vals_base, y_vals_base, color='blue', label=f'Penman-Monteith\nMean: {media_base:.2f}\nSD: {sd_base:.2f}', linewidth=1)  # Agrega la segunda lnea

plt.axvline(media_PT, color='red', linestyle='dashed', linewidth=1)
plt.axvline(media_base, color='blue', linestyle='dashed', linewidth=1)

plt.xlabel('Valores')
plt.ylabel('Densidad de Probabilidad')
plt.title('Grfico de Densidad de las Variables')

# Habilita las etiquetas de leyenda en mltiples lneas
plt.legend(loc='upper right')

plt.show()
```
